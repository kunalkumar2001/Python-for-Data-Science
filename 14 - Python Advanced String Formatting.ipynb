```python
def welcome_message(name):
    return ("Hello {}, Welcome to the community".format(name))
```

```python
welcome_message("Kunal")
```

```python
def welcome_message(name = "Kunal", role = "Data Science"):
    return ("Hello {} and role for {}, Welcome to the community".format(name,role))
```

```python
print(welcome_message())
```

```python
print(welcome_message("Amit", "Machine Learning"))
```

```python
name = "Kunal"
role = "Data Science"
print(f"Hello,{name},and role for {role}")
```
