# 'is' vs '=='

## == ‚Üí Equality operator

- Checks value equality
- üëâ ‚ÄúDo these two objects have the same value?‚Äù

```python
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)   
```

- ‚úî Values are the same.

# is ‚Üí Identity operator

- Checks memory identity
- üëâ ‚ÄúDo these two variables point to the same object in memory?‚Äù

```python
print(a is b)  
```

```python
print(id(a))
print(id(b))
```

- ‚ùå Different objects in memory.

### Python may reuse (cache) objects as an optimization.

- Small integers (-5 to 256) are always cached

- Larger integers may or may not be cached

- This depends on:

    - Python version

    - Compiler optimizations

    - Execution context (script, REPL, notebook)

```python
x = 1000
y = 1000

print(x == y)  
print(x is y)   
```

```python
print(id(x))
print(id(y))
```

```python
x = 10
y = 10

print(x is y) 
```

- üëâ Python interns small integers (typically -5 to 256) for performance.

```python
s1 = "python"
s2 = "python"

print(s1 == s2)  
print(s1 is s2)  
```

```python
s1 = "".join(["py", "thon"])
s2 = "python"

print(s1 == s2)  
print(s1 is s2)  
```
