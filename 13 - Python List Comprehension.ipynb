# Python List Comprehension
- List Comprehension provide a concise way to a create lists. It consists of brackets contaningan expression followed by a for clause, then zero or more for or if clauses. The expressions can be anything, meaning you can put in all kindsof objects in list.

## Syntax:-
- newlist = [expression for item in iterable if condition == True]

```python
lst1 = []
def lst_square(lst):
    for i in lst:
        lst1.append(i*i)
    return lst1
```

```python
lst_square([1,2,3,4,5,6,7,8,9])
```

```python
lst = [1,2,3,4,5,6,7,8,9]
```

```python
#List Comprehension
[i*i for i in lst]
```

```python
# if condition
[i*i for i in lst if i%2 == 0]
```

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = []
for x in fruits:
    if "a" in x:
        newlist.append(x)
```

```python
print(newlist)
```

### With list comprehension you can do all that with only one line of code:-

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = [x for x in fruits if "a" in x]
print(newlist)
```

```python
newlist = [x for x in fruits if x != "apple"]
```

```python
print(newlist)
```

```python
lst= [1,2,3,4,5,6,7,8,9]
lst1 = []
for i in lst:
    lst1.append(i*i)
```

```python
print(lst1)
```

```python
newlist = [x for x in range(10) if x < 5]
```

```python
newlist = [x.upper() for x in fruits]
```

```python
print(newlist)
```

```python
newlist = ['hello' for x in fruits]
print(newlist)
```

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]

newlist = [x if x != "banana" else "orange" for x in fruits]

print(newlist)
```
