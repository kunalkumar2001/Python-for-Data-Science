## How does Eval work
1. Parse python Expression
2. compile into a byte code
3. Evaluate the python expression
4. It will return the result

```python
var=compile("5*5","<string>","eval")
```

```python
var
```

```python
eval(var)
```

```python
### GLobals
x=10
eval("x+50+x**2",{"x":x})
```

```python
x=100
z=100
eval("x+z+w",{"x":x,"z":100,"w":1000})
```

```python
x=100
z=100
y=10000
eval("x+z+y")
```

```python
eval("a+b+c",{},{"a":100,"b":200,"c":1000})
```

```python
a=100
b=200
eval("a<b")
```

```python
x = 100
y = 100
result = eval("x + y", {}, {"x": x, "y": y})
print(result)
```

```python
x=100
y=100
eval("x+y",{},{},{"x":x,"y":y}) # here global and local values are empty
```
