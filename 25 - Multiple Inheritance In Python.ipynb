# Multiple Inheritance In Python

```python
class A:
    def method1(self):
        print("A class method is called")
```

```python
class B(A):
    def method1(self):
        print("B class method is called")
        
    def method2(self):
        print("B class2 method is called")
```

```python
class C(A):
    def method1(self):
        print("C class method is called")
```

```python
class D(B,C):
    def method1(self):
        print("D class method is called")
        
        C.method1(self)
        B.method1(self)
        A.method1(self)
```

```python
d = D()
```

```python
d.method1()
```

```python
d.method2()
```
