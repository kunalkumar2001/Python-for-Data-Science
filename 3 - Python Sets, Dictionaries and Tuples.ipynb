## Sets:-
- Sets are used to store multiple items in single variable.
- A set is collection which is unorderd, unchangeable, and unidexed.
- Set cannot have two items with same values.
- Python's set class represents the mathematical notion of a set.This is based on a data structure known as a hash table

```python
## Defining an empy set

set_var= set()
print(set_var)
print(type(set_var))
```

```python
# Duplicate not allowed
set1 = {"apple","mango","cherry","apple"}
print(set1)
```

### Access Set Items:-

- It cannot access itms in a set by referring to an index or key.
- But you can loop through the set items using a 'for' loop or ask if a specified value is present in a set, by using the 'in' keyword.

```python
#Example
set1 = {"apple","mango","cherry",}
for x in set1:
    print(x)
```

```python
set1 = {"apple","mango","cherry"}
print("mango" in set1)
```

### Change Items:-
- Once a set is created, you cannot change its items, but you can add new items.

### Add Items:-
- To add one item to a set use the add() method.

```python
set1 = {"apple","mango","cherry"}
set1.add("banana")
print(set1)
```

- To add items from another set into the current set, use the update() method.

```python
set1 = {"apple"}
set2 = {"banana"}
set1.update(set2)
print(set1
     )
```

#### Add any Iterable:-
- The object in the update() method does not have to be a set, it can be any iterable object(sets, tuples, list, dictionaries).

```python
set1 = {"apple","mango","cherry"}
list1 = ["kiwi","orange"]
set1.update(list1)
print(set1)
```

### Remove Item:-
- Use the remove() ot discard() method.


```python
set1 = {"apple","mango","cherry"}
set1.remove("apple")
print(set1)
```

```python
set1 = {"apple","mango","cherry"}
set1.discard("mango")
print(set1)
```

- Pop() method remove an item, but this method will remove a random item, so you cannot sure what item that gets removed.

- The del keywoed will delete the set complete.

### Joint Sets:-
- The union() method returns  a new set with all itms from both sets.

```python
set1 = {"apple","mango","cherry"}
set2 = {1,2,3}
set3 = set1.union(set2)
print(set3)
```

- The Intersection_update() method will keep only common items that are present in both sides.

```python
set1 = {"apple","mango","cherry"}
set2 = {"orange","mango","kiwi"}
set1.intersection_update(set2)
print(set1)
```

```python
set1 = {"apple","mango","cherry"}
set2 = {"orange","mango","kiwi"}
set3= set1.intersection(set2)
print(set3)
```

### Accessing Keys:-
- You can access the items of a dictionary by referring to its key name, inside square brackets[].

```python
dict1 = {"Brand":"Ford","Electric":"False","Year":1964,"Colors":["red","White"]}
x = dict1["Brand"]
print(x)
```

```python
# There is also a method called get() that will give you same result.
dict1 = {"Brand":"Ford","Electric":"False","Year":1964,"Colors":["red","White"]}
x = dict1.get("Brand")
```

```python
print(x)
```

### Get Keys:-

```python
dict1 = {"Brand":"Ford","Electric":"False","Year":1964,"Colors":["red","White"]}
x = dict1.keys()
print(x)
```

### Removing Items:-

```python
# The pop() method removes the item with the specified key name.
dict1 = {"Brand":"Ford","Electric":"False","Year":1964,"Colors":["red","White"]}
dict1.pop("Brand")
print(dict1)
```

```python
# The del() method also removes the item with the specified key name.
dict1 = {"Brand":"Ford","Electric":"False","Year":1964,"Colors":["red","White"]}
del dict1["Brand"]
print(dict1)
```

```python
# The popitem() method removes the last inserted item.
dict1 = {"Brand":"Ford","Electric":"False","Year":1964,"Colors":["red","White"]}
dict1.popitem()
print(dict1)
```

```python
# The clear() method empties the dictionary.
# The pop() method removes the item with the specified key name.
dict1 = {"Brand":"Ford","Electric":"False","Year":1964,"Colors":["red","White"]}
dict1.clear()
print(dict1)
```

### Loop Method:-

```python
# Print all keys name in the dictionary one by one.
dict1 = {"Brand":"Ford","Electric":"False","Year":1964,"Colors":["red","White"]}
for x in dict1:
    print(x)
```

```python
dict1 = {"Brand":"Ford","Electric":"False","Year":1964,"Colors":["red","White"]}
for x in dict1.keys():
    print(x)
```

```python
# Print all values in the dictionary one by one.
dict1 = {"Brand":"Ford","Electric":"False","Year":1964,"Colors":["red","White"]}
for x in dict1:
    print(dict1[x])
```

```python
for x in dict1.values():
    print(x)
```

```python
for x,y in dict1.items():
    print(x,y)
```

### Copy Dictionary:-

```python
dict1 = {"Brand":"Ford","Electric":"False","Year":1964,"Colors":["red","White"]}
my_dict = dict1.copy()
print(my_dict)
```

### Nested Dictionaries:-
- A dictionary can contain dictionaries is called Nested dictionaries.

```python
my_family = {"child1":{"Name":"Emil","Year":2004},
            "child2":{"Name":"Thomas","Year":2007},
            "child3":{"Name":"Linus","Year":2011}}
print(my_family)
```

```python
child1 = {"Name":"Emil","Year":2004}
child2 = {"Name":"Thomas","Year":2007}
child3 = {"Name":"Linus","Year":2011}

my_family = {"child1":child1,"child2":child2,"child3":child3}
print(my_family)
```

```python
# Access items in Nested Dictionaries:-
print(my_family["child2"]["Name"])
```
