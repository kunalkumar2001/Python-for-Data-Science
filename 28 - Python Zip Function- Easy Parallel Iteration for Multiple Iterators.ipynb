```python
lst1 = ["Kunal","Kumar","Data Science"]
lst2 = ["a","b","c"]
```

```python
for i in lst1:
    print(i)
```

```python
output = zip(lst1,lst2)
```

```python
output
```

```python
list(output)
```

```python
# Zip object is an iterator
next(output)
```

```python
for i in output:
    print(i)
```

```python
output = zip(lst1,lst2)
for i,j in output:
    print(i,j)
```

```python
lst1 = ["Kunal","Kumar","Data Science"]
lst2 = ["a","b","c"]
lst3 = [1,2,3]
```

```python
output = zip(lst1, lst2, lst3)
```

```python
list(output)
```

```python
for i,j,k in output:
    print(i,j,k)
```
