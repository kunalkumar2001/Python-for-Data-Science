## Boolean Variables

Boolean values are the two constant objects False and True.

They are used to represent truth values (other values can also be considered false or true).

In numeric contexts (for example, when used as the argument to an arithmetic operator), they behave like the integers 0 and 1, respectively.

The built-in function bool() can be used to cast any value to a Boolean, if the value can be interpreted as a truth value

They are written as False and True, respectively.

## Lists:-
- Lists are used to store multiple items in single variables.
- Lists are one of 4-built- in data types in Python used to store collection of data
- list is a data structure in Python that is a mutable, or changeable, ordered sequence of elements and allow duplicates values.
- Each element or value that is inside of a list is called an item.
- List items can be of any data types.
- Lists are created using Square brackets [].

### Change Item Value:-
- To change the value of a specific item refer to the index numbers

```python
lst = ["mango","banana","cherry"]
lst[1]= "apple"
print(lst)
```

```python
# Change a range of Items Value
lst = ["apple","banana","cherry"]
lst[1:2] = ["mango","kiwi"]
print(lst)
```

```python
lst = ["apple","banana","cherry"]
lst[1:3] = ["mango"]
print(lst)
```

### Append Items:-
- To add an item to the end of the list use the append() method.


```python
lst = ["apple","banana","cherry"]
lst.append("mango")
print(lst)
```

```python
lst = [1,2,3]
```

```python
lst.append([4,5])
print(lst)
```

### Insert Items:-
- To insert a new list item without replacing 

```python
lst = ["apple","banana","cherry"]
lst.insert(2,"mango")
print(lst)
```

### Extend List:-
- To append elements from another list to the current list, use the extend() method.

```python
lst1 = ["apple","banana","cherry"]
lst2 = ["mango","kiwi","papaya"]
```

```python
lst1.extend(lst2)
print(lst1)
```

### Remove Specified Item:-
- The remove() method removes the specified item.

```python
lst = ["apple","banana","cherry"]
lst.remove("banana")
print(lst)
```

### Pop Method:-
- THis method used to remove the specific index.

```python
lst = ["apple","banana","cherry"]
lst.pop(1)
```

```python
print(lst)
```

```python
# if you do not specify the index, the pop() method removes the last item.
lst = ["apple","banana","cherry"]
lst.pop()
print(lst)
```

### Delete Method:-
- The del keyword also removes the specific index.

```python
lst = ["apple", "banana", "cherry"]
del lst[0]
print(lst)
```

```python
lst = ["apple", "banana", "cherry"]
del lst
print(lst) # Error because list are empty.
```

### Clear List:-
- The Clear() method empties the list.
- The list still remains, but it has no content.

```python
lst = ["apple", "banana", "cherry"]
lst.clear()
print(lst)
```

### Loop through a list:-
- It can llop through the list items by using a 'for' loop.


```python
lst = ["apple", "banana", "cherry"]
for x in lst:
    print(x)
```

```python
lst = ["apple", "banana", "cherry"]
for i in range (len(lst)):
    print(lst[i])
```

```python
# Using a while loop:-
lst = ["apple", "banana", "cherry"]
i=0
while i< len(lst):
    print(lst[i])
    i=i+1
```

### Sort list Alphanumerically:-
- List objects have a sort() method that will sort the list alphanumerically ascending order.

```python
lst = ["orange","mango","apple","kiwi"]
lst.sort()
print(lst)
```

```python
lst1 = [20,10,50,30]
lst1.sort()
print(lst1)
```

### Sort Descending:-
- To sort descending, use the keyword argument reverse= True.

```python
lst = ["orange","mango","apple","kiwi"]
lst.sort(reverse=True)
print(lst)
```

```python
lst1 = [20,10,50,30]
lst1.sort(reverse=True)
print(lst1)
```

```python
# Case Sensitive Order
lst = ["orange","Mango","apple","Kiwi"]
lst.sort()
print(lst)
```

```python
lst = ["orange","Mango","apple","Kiwi"]
lst.reverse()
print(lst)
```

```python
# Copy a list:-
lst = ["orange","mango","apple","kiwi"]
lst1 = lst.copy()
print(lst1)
```

```python
lst1 = ["orange","mango","apple","kiwi"]
lst2 = [1,2,3,4]
lst = lst1 + lst2
print(lst)
```
